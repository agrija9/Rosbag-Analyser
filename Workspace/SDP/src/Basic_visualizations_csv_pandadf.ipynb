```python
from dateutil.parser import parse 
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
plt.rcParams.update({'figure.figsize': (10, 7), 'figure.dpi': 120})
```

```python
path_to_file = "/home/shravanthi/Downloads/sem2/Rosbag_Content.csv"
df = pd.read_csv(path_to_file,encoding = 'utf-8',parse_dates =['Time'])

df.head()
```

```python
df1 = df.head()
def plot_df(df, x, y, title="", xlabel='Time', ylabel='Topic', dpi=100):
    plt.figure(figsize=(16,8), dpi=dpi)
    plt.plot(x, y, color='tab:red')
    plt.gca().set(title=title, xlabel=xlabel, ylabel=ylabel)
    plt.show()

plot_df(df1, x=df.Time, y=df.Topic, title='Plot')    
```
